**Respuestas**

1. **Apache Spark** es una plataforma de procesamiento distribuido diseñada para manejar grandes volúmenes de datos a alta velocidad. A diferencia de Pandas, que opera en un solo nodo y está limitado por la memoria disponible, Spark permite procesar datos de manera distribuida en múltiples nodos, aprovechando la memoria distribuida y la paralelización. Además, Spark puede trabajar tanto en memoria como en disco, lo que le da mayor flexibilidad y eficiencia para manejar datasets de tamaño considerable. Esto lo convierte en una solución escalable y rápida frente a Pandas.

2. Un **RDD** en Spark es una colección inmutable de objetos distribuidos en el clúster que permite la tolerancia a fallos y el procesamiento paralelo. Una de sus principales características es su capacidad de ser reconstruido automáticamente en caso de fallos en los nodos, ya que almacena información sobre cómo fue creado. En comparación con los dataframes, que tienen un esquema estructurado y optimizaciones para SQL, los RDD son más flexibles, aunque menos eficientes para ciertos casos de uso. Ambos se usan en el contexto del procesamiento distribuido.

3. Las **transformaciones** en Spark son operaciones perezosas que crean nuevos RDDs o DataFrames, como `map()`, `filter()`, pero no se ejecutan hasta que se realiza una acción. Las **acciones**, como `collect()`, `count()`, disparan el cálculo real y devuelven un resultado al controlador. Internamente, cuando se ejecuta una acción, Spark construye un grafo dirigido acíclico (DAG) de las transformaciones para optimizar la ejecución y dividir el trabajo entre los nodos, mejorando el rendimiento del clúster.

4. El **particionamiento** en Spark es crucial porque determina cómo se distribuyen los datos en los diferentes nodos del clúster para su procesamiento. Un particionamiento eficiente garantiza que el procesamiento se realice en paralelo de manera equilibrada, evitando sobrecargar algunos nodos y dejando otros ociosos. Un mal particionamiento puede causar "cuellos de botella" o una sobrecarga de comunicación entre nodos, lo que afecta negativamente el rendimiento general del trabajo. En resumen, un particionamiento adecuado es clave para maximizar la eficiencia de los recursos.

5. El **Spark Driver** es el componente que coordina todo el proceso de ejecución de un trabajo en Spark. Planifica, envía y rastrea las tareas en los diferentes nodos del clúster. Los **Spark Executors**, por otro lado, son los responsables de ejecutar esas tareas en los nodos, administrar la memoria y devolver los resultados al Driver. Cada nodo en el clúster puede tener múltiples ejecutores, lo que permite paralelizar el trabajo y manejar grandes volúmenes de datos de manera eficiente.

6. El **Catalyst Optimizer** es el componente encargado de optimizar las consultas SQL en Spark. Su principal función es analizar y transformar las consultas para generar un plan de ejecución eficiente. Lo hace mediante un proceso de reescritura y optimización basado en reglas y patrones. El optimizador aplica técnicas como la proyección y el filtrado de columnas, la reordenación de operaciones, entre otras, para minimizar el tiempo de procesamiento y mejorar el uso de los recursos del clúster.

````python
from pyspark.sql.types import BooleanType
def cast_columns(df, cols_types):
    """
    Cambia el tipo de múltiples columnas en un DataFrame de Spark.

    Parámetros:
    - df: DataFrame de Spark.
    - cols_types: Diccionario con nombres de columnas como claves y tipos de datos de Spark como valores.

    Retorna:
    - DataFrame de Spark con tipos de columnas modificados.
    """
    for col_name, new_type in cols_types.items():
        df = df.withColumn(col_name, df[col_name].cast(new_type))
    return df

# Ejemplo de uso
cols_to_cast = {"Quantity": IntegerType(), "Customer Age": IntegerType(),
                "Is An Alien": BooleanType(), "Account Age Days": IntegerType(), "Transaction Hour": IntegerType()}
df = cast_columns(df, cols_to_cast)
df.printSchema()
````

Para continuar con el análisis, los especistas de Bodoque les gustaría tener nuevas variables disponibles. Tras las notas de la reunión usted llega a la conclusión de que tiene que realizar las siguientes tareas (con el dataset preprocesado de la seccion anterior):


1.   Agregar una columna llamada "Transaction bp" con el **monto total** de la transacción en bodoque pesos. Se considera que $x$ dólares equivalen a $log(48+|x^{36}|)$ bodoque pesos.
2.   Crear una columna llamada "Transaction Month" con el mes en que se realiza una transacción.
3.   Crear la variable *Type of purchase* según la catidad de unidades vendidas de acuerdo a las siguientes categorías.
  * Compra minorista: 4 productos o menos.
  * Compra mayorista: 5 produtos o más.
4. Imprima los registros de compras hechas por alienígenas en el comecio mayorista.  Utilice `.filter()`.
5. Cuente la cantidad de compras realizadas por humanos y la cantidad de compras realizadas por alienígenas. Utilice `.groupby()`.
6. Muestre una tabla con la recaudación promedio por transacción para cada método de pago, tanto para humanos como alienígenas. Utilice `pivot()`.

````python
# Escriba su respuesta aquí
from pyspark.sql.functions import col, log, abs, month, when, avg

# 1. Agregar columna "Transaction bp"
df = df.withColumn("Transaction bp", log(48 + abs(col("Transaction Amount") / 36)))

# 2. Agregar columna "Transaction Month"
df = df.withColumn("Transaction Month", month(col("Transaction Date")))

# 3. Crear la variable "Type of purchase"
df = df.withColumn(
    "Type of purchase",
    when(col("Quantity") <= 4, "Compra minorista")
    .otherwise("Compra mayorista")
)

# 4. Filtrar compras mayoristas realizadas por alienígenas
alien_mayoristas = df.filter((col("Is An Alien") == True) & (col("Quantity") >= 5))
alien_mayoristas.show()

# 5. Contar compras realizadas por humanos y alienígenas
count_humans_aliens = df.groupBy("Is An Alien").count()
count_humans_aliens.show()

# 6. Recaudación promedio por transacción para cada método de pago y tipo de cliente
avg_recaudacion = df.groupBy("Payment Method", "Is An Alien")\
    .agg(avg("Transaction Amount").alias("Avg Transaction Amount"))\
    .groupBy("Payment Method")\
    .pivot("Is An Alien", ["True", "False"])\
    .agg(avg("Avg Transaction Amount"))

avg_recaudacion.show()
````

El equipo de Bodoque e-shop ha solicitado que los datos estén disponibles en una tabla SQL consultable. Además, están interesados en aprovechar las funciones de ventana en SQL para análisis avanzados. Las funciones de ventana permiten realizar cálculos sobre un conjunto de filas que están relacionadas con la fila actual. Por ejemplo, UNBOUNDED PRECEDING se usa para indicar que el rango de la función de ventana comienza desde la primera fila de la partición o del conjunto de resultados, lo cual es útil para calcular sumas acumulativas hasta la fila actual. Las variaciones comunes de este uso incluyen:

- `UNBOUNDED PRECEDING` to `CURRENT ROW`: Calcula desde el inicio de la partición hasta la fila actual.
- `UNBOUNDED PRECEDING` to `UNBOUNDED FOLLOWING`: Cubre todas las filas dentro de la partición.
- `VALUE PRECEDING` to `VALUE FOLLOWING`: Establece un rango específico basado en valores antes y después de la fila actual.

<center>
<img src="https://learnsql.com/blog/sql-window-functions-rows-clause/1.png" width=500 />

Ejemplo de uso en SQL:

```sql
STAT(COL1_NAME) OVER (PARTITION BY COL2_NAME ORDER BY COL3_NAME ROWS BETWEEN X PRECEDING AND CURRENT ROW)
```

````python
# Código Aquí
from pyspark.sql import Window
from pyspark.sql.functions import max, col
from pyspark.sql.functions import current_date, date_sub

# Definir una ventana particionada por cliente y ordenada por fecha de transacción
window_spec = Window.partitionBy("Customer ID").orderBy("Transaction Date").rowsBetween(Window.unboundedPreceding, Window.currentRow)

# Filtrar transacciones de los últimos 3 meses
df_filtered = df.filter(col("Transaction Date") >= date_sub(current_date(), 90))

# Calcular la transacción más alta por cliente dentro de los últimos 3 meses
df_with_max = df_filtered.withColumn("Max_Transaction", max("Transaction Amount").over(window_spec))

# Mostrar resultados
df_with_max.select("Customer ID", "Max_Transaction").orderBy("Customer ID").show()
````
